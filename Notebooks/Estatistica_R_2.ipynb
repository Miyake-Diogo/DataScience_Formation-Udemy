## Regressão logistica, conjunto de dados de eleição, será usado.

```r
eleicao = read.csv(file.choose(), sep=";", header=T)#arquivo candidatos.csv
eleicao
```

```r
fix(eleicao)
```

```r
plot(eleicao$DESPESAS, eleicao$SITUACAO)
```

```r
summary(eleicao)
```

```r
cor(eleicao$DESPESAS, eleicao$SITUACAO)
```

```r
#criação do modelo
modelo_eleicao = glm(SITUACAO ~ DESPESAS, data = eleicao, family = "binomial") #regressão linear de outro tipo, que é passada como parâmetro
summary(modelo_eleicao)
```

```r
plot(eleicao$DESPESAS, eleicao$SITUACAO, col="red", pch=20) # cor verelha e preenchimento
points(eleicao$DESPESAS, modelo_eleicao$fitted, pch = 4) # mpontos de ajuste da probabilidade de ser eleito
```

```r
prever_eleicao = read.csv(file.choose(), sep=";", header=T) #arquivo novoscandidatos.csv
prever_eleicao
```

```r
fix(prever_eleicao)
```

```r
prever_eleicao$RESULT = predict(modelo_eleicao, newdata=prever_eleicao, type = "response") # criação de coluna e predição de cada candidato ser eleito
```

```r
prever_eleicao$RESULT
```
