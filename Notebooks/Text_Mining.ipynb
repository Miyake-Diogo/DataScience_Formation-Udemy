```r
stopwords("portuguese") #retorna palavras sem valor semântico
```

```r
corpus = tm_map(corpus,removeWords,stopwords("english"))
```

```r
as.character(corpus[[3]])
```

Nota se a diferença do corpus[[3]] original para o sem os stopwords<br/>

```r
#'\'NORTHERN TELECOM LTD &lt;NT> DECLARES STOCK SPLIT Two-for-one stock split'
#'    Pay May 12'
#'    Note: split is subject to approval of shareholders at April'
#'23 annual meeting.'
#'    Company also said it will increase dividend on post-split'
#'shares to six cts from five cts.'
#' Reuter'
#'&#3;\',0'
```

```r
corpus = tm_map(corpus, stripWhitespace) #remove os espaços
corpus = tm_map(corpus, removePunctuation) #remove a pontuação
corpus = tm_map(corpus, removeNumbers) # remove numeros
```

```r
as.character(corpus[[3]])
```

### Matriz de termos mais frequentes
TermDocumentMatrix e DocumentTermMatrix

```r
frequencia = TermDocumentMatrix(corpus)
```

```r
matriz = as.matrix(frequencia) #criação de uma matriz
matriz = sort(rowSums(matriz), decreasing=T) #ordenar a matriz de forma decrescente
matriz = data.frame(word=names(matriz), frequencia=matriz) #criar um data frame da matriz
```
